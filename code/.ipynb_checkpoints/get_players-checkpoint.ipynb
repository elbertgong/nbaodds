```python
from bs4 import BeautifulSoup
import requests
import csv
import numpy as np
import pandas as pd
```

```python
actives = []
def filt(player):
    if player.find('strong') is not None:
        return 'basketball-reference.com' + player.find('a')['href']
```

```python
for i in range(ord('a'), ord('z')+1):
    if i == 120:
        continue
    bballurl = "https://www.basketball-reference.com/players/%c/" % chr(i)
    result = requests.get(bballurl)
    # result.headers
    c = result.content
    soup = BeautifulSoup(c, 'html.parser')
    t = soup.table.tbody.find_all('tr')
    mapped = list(map(filt,t)) 
    actives.extend([x for x in mapped if x is not None])
```

```python
actives
```

```python
with open('../data/active_player.csv', 'w') as csvfile:
    writer = csv.writer(csvfile, delimiter=' ',
                            quotechar='|', quoting=csv.QUOTE_MINIMAL)
    for player in actives:
        writer.writerow([player])
```
